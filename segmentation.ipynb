```python
features = ['Age', 'Income','Total_spending', 'NumWebPurchases', 'NumStorePurchases','NumWebVisitsMonth','Recency' ]
```

```python
X = df[features].copy()
```

```python
X
```

PREPROCESSING


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X) 
```

```python
X_scaled
```

```python
#Finding the optimal number of clusters
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(2,15):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2,15),wcss,marker = "o")
plt.title('Elbow Method for Optimal K')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
scores = []

for k in range(2,15):
    kmeans = KMeans(n_clusters=k)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    scores.append(score)

print(scores)
```

```python
plt.plot(range(2,15), scores, marker ="o")
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhoette Method for Optimal K')
plt.show()
```

```python
k_values = range(2, 15)

fig, ax1 = plt.subplots(figsize=(8, 5))

# ---- 1. Y-ekseni: WCSS (Elbow) ----
ax1.plot(k_values, wcss, marker='o', linestyle='-', label='WCSS')
ax1.set_xlabel('k (number of clusters)')
ax1.set_ylabel('WCSS', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# ---- 2. Y-ekseni: Silhouette ----
ax2 = ax1.twinx()
ax2.plot(k_values, scores, marker='s', linestyle='--', label='Silhouette',color ="orange")
ax2.set_ylabel('Silhouette Score', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

plt.title('Elbow (WCSS) and Silhouette Scores vs k')
fig.tight_layout()
plt.show()
```

Choosed k =6 

```python
kmeans = KMeans(n_clusters= 6)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```
